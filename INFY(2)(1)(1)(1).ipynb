```python
temp_df.drop(["Dividends", "Stock Splits"], axis = 1)
```

```python
df = temp_df.drop(["Dividends", "Stock Splits"], axis = 1)
```

```python
df = df.loc['2022-01-01': '2023-01-01']
df
```

```python
training_set = df.iloc[:, 1: 2].values
print(training_set)
print(training_set.shape)
```

```python
from sklearn.preprocessing import MinMaxScaler
scaler = MinMaxScaler(feature_range = (0, 1))
scaled_training_set = scaler.fit_transform(training_set)
scaled_training_set
```

```python
X_Train = []
y_Train = []
for i in range(60, 251):
    X_Train.append(scaled_training_set[i - 60 : i, 0])
    y_Train.append(scaled_training_set[i, 0])
X_Train = np.array(X_Train)
y_Train = np.array(y_Train)
print(X_Train.shape)
print(y_Train.shape)
```

```python
X_Train = np.reshape(X_Train, (X_Train.shape[0], X_Train.shape[1], 1))
X_Train.shape
```

```python
from keras.models import Sequential
from keras.layers import LSTM
from keras.layers import Dense
from keras.layers import Dropout
```

```python
pip install tensorflow
```

```python
regressor = Sequential()
regressor.add(LSTM(units = 50, return_sequences = True, input_shape = (X_Train.shape[1], 1)))
regressor.add(Dropout(0.2))
regressor.add(LSTM(units = 50, return_sequences = True))
regressor.add(Dropout(0.2))
# regressor.add(LSTM(units = 50, return_sequences = True))
# regressor.add(Dropout(0.2))
regressor.add(LSTM(units = 50))
regressor.add(Dropout(0.2))
regressor.add(Dense(units = 1))
```

```python
regressor.compile(optimizer = 'adam', loss = 'mean_squared_error')
regressor.fit(X_Train, y_Train, epochs = 100, batch_size = 32)
```

```python
hist
```

```python
dataset_test = temp_df.loc['2023-01-01':]
dataset_test
actual_stock_price = dataset_test.iloc[:, 1:2].values
print(actual_stock_price)
print(actual_stock_price.shape)
```

```python
dataset_total = pd.concat((df['Open'], dataset_test['Open']), axis = 0)
print('dataset')
print(dataset_total)
print('-----dataset')

inputs = dataset_total[len(dataset_total) - len(dataset_test)-60:].values

print('inputs')

print(inputs)
print('------------------------------inputs')

inputs = inputs.reshape(-1, 1)
inputs = scaler.transform(inputs)
X_test = []
for i in range(60, 121):
    X_test.append(inputs[i - 60:i, 0])
print('X_test')

print(X_test)
print('End ------ X_test')

X_test = np.array(X_test)
X_test
X_test = np.reshape(X_test, (X_test.shape[0], X_test.shape[1], 1))
```

```python
predicted_stock_price = regressor.predict(X_test)
predicted_stock_price = scaler.inverse_transform(predicted_stock_price)
```

```python
plt.plot(actual_stock_price, color = 'red', label = 'Actual infosys stock price')
plt.plot(predicted_stock_price, color = 'blue', label = 'Predicted infosys stock price')
plt.xlabel('Time')
plt.ylabel('Infosys stock price')
plt.legend()
```
